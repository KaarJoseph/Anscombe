### 1. Cálculo de estadísticas

En esta sección calculamos las estadísticas básicas de cada grupo: media y desviación estándar. Estos valores estadísticos se utilizan para demostrar que, aunque cada grupo comparte estadísticas similares, sus distribuciones gráficas son notablemente diferentes.

```python
import numpy as np

# Ejemplo de arreglo
arreglo = [1.015, 0.990, 1.013, 1.001, 0.986]

# Media
media = np.mean(arreglo)

# Desviación estándar
desviacion_estandar = np.std(arreglo)

print("Media:", media)
print("Desviación Estándar:", desviacion_estandar)
```

El análisis destaca que los valores de cada grupo (como las medias y desviaciones) son notablemente similares entre sí, lo cual es clave en el cuarteto de Anscombe. Sin embargo, estos números, por sí solos, no reflejan las diferencias gráficas que se evidenciarán en la visualización.

### 3. Visualización de los Datos de Anscombe con Gráficos de Dispersión

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo CSV (ya lo tienes cargado)
archivo_csv = 'df_anscombe.csv'
df = pd.read_csv(archivo_csv)

# Establecer el estilo de Seaborn
sns.set(style="whitegrid")

# Crear la figura y los subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle("Anscombe's quartet")

# Gráfica para el Grupo 1
sns.scatterplot(ax=axs[0, 0], x='x', y='y', data=df[df['group'] == 1])
axs[0, 0].set_title('Group 1')

# Gráfica para el Grupo 2
sns.scatterplot(ax=axs[0, 1], x='x', y='y', data=df[df['group'] == 2])
axs[0, 1].set_title('Group 2')

# Gráfica para el Grupo 3
sns.scatterplot(ax=axs[1, 0], x='x', y='y', data=df[df['group'] == 3])
axs[1, 0].set_title('Group 3')

# Gráfica para el Grupo 4
sns.scatterplot(ax=axs[1, 1], x='x', y='y', data=df[df['group'] == 4])
axs[1, 1].set_title('Group 4')

# Ajustar el layout
plt.tight_layout()
plt.subplots_adjust(top=0.9)
plt.show()
```

Estos gráficos muestran que cada grupo tiene una relación única entre \(x\) e \(y\), revelando patrones de distribución, alineación lineal y posibles outliers. Este análisis visual refuerza que las estadísticas por sí solas no capturan la naturaleza completa de los datos.

### 4. Diagramas de Caja (Box Plots) para Comparación Adicional

```python
# Establecer el estilo de Seaborn
sns.set(style="whitegrid")

# Crear la figura y los subplots para los boxplots
fig, axs = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle("Boxplots por Grupo con Outliers")

# Boxplot para el Grupo 1 (ajustando 'whis')
sns.boxplot(ax=axs[0, 0], x='group', y='y', data=df[df['group'] == 1], whis=1.0)
axs[0, 0].set_title('Boxplot - Grupo 1')

# Boxplot para el Grupo 2 (ajustando 'whis')
sns.boxplot(ax=axs[0, 1], x='group', y='y', data=df[df['group'] == 2], whis=1.0)
axs[0, 1].set_title('Boxplot - Grupo 2')

# Boxplot para el Grupo 3 (ajustando 'whis')
sns.boxplot(ax=axs[1, 0], x='group', y='y', data=df[df['group'] == 3], whis=1.0)
axs[1, 0].set_title('Boxplot - Grupo 3')

# Boxplot para el Grupo 4 (ajustando 'whis')
sns.boxplot(ax=axs[1, 1], x='group', y='y', data=df[df['group'] == 4], whis=1.0)
axs[1, 1].set_title('Boxplot - Grupo 4')

# Ajustar el layout
plt.tight_layout()
plt.subplots_adjust(top=0.9)
plt.show()
```

Los box plots destacan las diferencias en la distribución y presencia de outliers en cada grupo, reforzando la idea de que, aunque los grupos compartan estadísticas similares, sus distribuciones tienen variaciones que se pueden observar en esta visualización.
